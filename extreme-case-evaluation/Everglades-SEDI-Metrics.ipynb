```python
## For NP205 stage ###

import pandas as pd
import numpy as np  # Make sure to import numpy

# Assuming df is your DataFrame
unique_id = 'NP205_stage'

# Filter dataset for the unique_id
df_filtered = df[df['unique_id'] == unique_id]

# Extract actual and predicted values
y_true = df_filtered['y']
predicted_columns = ['NBEATS', 'Informer', 'DLinear', 'PatchTST', 'iTransformer', 'NLinear', 'RMoK', 
                     'TimeMixer', 'TSMixer', 'TSMixerx', 'KAN', 'TimeGPT', 'Chronos', 'TimeLLM', 'timesfm', 'Morai', 'Timer']

# Initialize dictionary to store SEDI values
sedi_results = {}

# Define function to calculate SEDI score for a given threshold
def calculate_sedi(true_values, predicted_values, low_threshold, high_threshold):
    yp_low = np.percentile(true_values, low_threshold * 100)
    yp_up = np.percentile(true_values, high_threshold * 100)

    numerator = 0
    denominator = 0
    
    # Loop through each prediction and calculate SEDI
    for true_value, pred_value in zip(true_values, predicted_values):
        true_extreme_low = true_value < yp_low
        true_extreme_up = true_value > yp_up
        pred_extreme_low = pred_value < yp_low
        pred_extreme_up = pred_value > yp_up
        
        if true_extreme_low:
            denominator += 1
        if true_extreme_up:
            denominator += 1
        if true_extreme_low and pred_extreme_low:
            numerator += 1
        if true_extreme_up and pred_extreme_up:
            numerator += 1
    
    # Return SEDI score, avoiding division by zero
    return numerator / denominator if denominator != 0 else 0

# Define percentiles to calculate SEDI for
percentiles = [0.995, 0.90]

# Initialize a dictionary to store SEDI scores for each model and unique_id
sedi_scores_by_id = {}

# Calculate SEDI scores for each model at 99.5th and 90th percentiles
for model in predicted_columns:
    model_scores = []
    for p in percentiles:
        sedi_score = calculate_sedi(y_true, df_filtered[model], 1 - p, p)
        model_scores.append(sedi_score)
    sedi_scores_by_id[model] = model_scores

# Store the calculated SEDI scores for this unique_id
sedi_results[unique_id] = sedi_scores_by_id

# Convert to DataFrame for easy visualization
sedi_df1 = pd.DataFrame(sedi_results[unique_id], index=[f"99.5th ↑", f"90th ↑"])
print(sedi_df1)
```
