```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define station mappings for legend labels
model_labels = {
    'TimesFM': 'TimesFM',
    'TimeGPT': 'TimeGPT',
    'Timer'  :  'Timer',
    'Moirai'  :  'Moirai',
    'Chronos': 'Chronos',
}

# List of models
models = list(model_labels.keys())
```

```python
# Dictionary to store MAE values
mae_results = {model: [] for model in models}

# Manually input MAE values for each input length (25, 50, 75, 100, 125, 150)
manual_mae_values = {
    'TimesFM': [0.190, 0.172, 0.219, 0.186, 0.180, 0.172],
    'TimeGPT': [0.161, 0.151, 0.1512, 0.151, 0.151, 0.151],
    'Timer'  : [0.210, 0.198, 0.230, 0.209, 0.167, 0.164],
    'Moirai'  : [0.292, 0.239, 0.252, 0.243, 0.250, 0.240],
    'Chronos': [0.139, 0.122, 0.127, 0.107, 0.112, 0.108],
}
```

```python
# Populate mae_results with manual values
for model in models:
    mae_results[model] = manual_mae_values[model]
```

```python
# Convert to DataFrame for plotting
mae_df = pd.DataFrame(mae_results, index=[25, 50, 75, 100, 125, 150])

# Plot the results
plt.figure(figsize=(10, 6))
for model in models:
    plt.plot(mae_df.index, mae_df[model], marker='o', label=model_labels[model])  # Use mapped labels

plt.xlabel("Input Length (Days)", fontsize=19)
plt.ylabel("MAE", fontsize=21)
plt.ylim(0.05,0.35)
plt.xticks([25, 50, 75, 100, 125, 150])
plt.xticks(fontsize=19)
plt.yticks(fontsize=19)

# # Move the legend to the top
# # plt.legend(title="Models", loc="upper center", fontsize=14.5, 
#         title_fontsize=15, labelspacing=1.2, bbox_to_anchor=(0.5, 1.20), ncol=len(models))

# Move the legend to the top right corner
plt.legend(loc="upper right", fontsize=13, labelspacing=0.5)



plt.grid(True)
plt.show()
```
